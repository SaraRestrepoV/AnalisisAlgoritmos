Contains duplicate:

Valida si en una lista de número se encuentra algún dato repetido.

Solución óptima

Recorre e conjunto una sola vez porque utiliza un set que recuerda los números ya recorridos. Si el número se encuentra ya en el conjunto lo cataloga como repetido.

```python
numeros = [1, 2, 3, 1]

numeros_vistos = set()
encontro_duplicado = False
numero_duplicado = None

for numero in numeros:
    if numero in numeros_vistos:
        encontro_duplicado = True
        numero_duplicado = numero
        break
    numeros_vistos.add(numero)

print(encontro_duplicado)
print("Número duplicado:", numero_duplicado)
```
