STONES ON THE TABLE

La regla es que si dos piedras se encuentran juntas no pueden tener el mismo color. El programa calcula cuantas piedras se deben quitar
para que ninguna piedra quede al lado de otra que tiene el mismo color

Solución fuerza bruta:

Se prueban todas las combinaciones posibles de piedras a eliminar, empezando desde 0 eliminaciones hasta n.
la primera combinación que sea váida genera a respuesta

```python
from itertools import combinations

colores = "RRRRRRG"
numero_piedras = len(colores)

for cantidad_a_eliminar in range(numero_piedras + 1):
    
    for indices_a_eliminar in combinations(range(numero_piedras), cantidad_a_eliminar):
        
        fila_restante = [colores[i] for i in range(numero_piedras) if i not in indices_a_eliminar]
        
        es_valida = True
        for i in range(1, len(fila_restante)):
            if fila_restante[i] == fila_restante[i - 1]:
                es_valida = False
                break
        
        # Si encontramos una combinación válida, imprimimos y terminamos
        if es_valida:
            print(cantidad_a_eliminar)
            break
    else:
        continue
    break
```

Solución óptima: Soluciona el problema de la forma más eficiente posible

```python
colores = "RRG"
numero_piedras = len(colores)
piedras_a_eliminar = 0

for i in range(1, numero_piedras):
    if colores[i] == colores[i - 1]:
        piedras_a_eliminar += 1

print(piedras_a_eliminar)
```
